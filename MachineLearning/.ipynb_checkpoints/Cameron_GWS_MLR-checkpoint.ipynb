```python
import pandas as pd
import json
import io
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
file_path = '../Data/gws_cleaned_dataset.csv'
df = pd.read_csv(file_path)
```

```python
df
```

```python
df['regions'] = df['regions'].str[2:]
df['regions'] = df['regions'].str[:-2]
df
```

```python
df.nunique()
```

```python
is_primeurs = df.loc[df['is_primeurs'] == True]
is_primeurs
```

```python
regions_t = is_primeurs['regions'].drop_duplicates()
regions_t
```

```python
region_prim = df.loc[(df['regions'] == "Bordeaux") | (df['regions'] == 'Sud Ouest')]
region_prim
```

```python
region_prim['review_year'] = pd.DatetimeIndex(region_prim['date']).year
region_prim['wine_age'] = np.subtract(region_prim['review_year'], region_prim['vintage'])
region_prim
```

```python
from sklearn.preprocessing import LabelEncoder
y = region_prim.is_primeurs
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
encoded_y
```

```python
# from keras.utils import to_categorical

# # Step 2: One-hot encoding
# one_hot_y = to_categorical(encoded_y)
# one_hot_y
```

```python
# target = np.delete(one_hot_y, 1, axis=1)
region_prim['prem'] = encoded_y.tolist()
```

```python
# df_new = pd.concat([region_prim, pd.DataFrame(target)], axis=1)
```

```python
# region_prim
```

```python
# X = region_prim[["vintage", 'score','journalist_count', 'wine_age']]
# y = target.reshape(-1, 1)
# print(X.shape, y.shape)

X = region_prim[["vintage", 'prem', 'journalist_count', 'wine_age']]

# X = region_prim['prem'].values.reshape(-1, 1)
# X = region_prim['journalist_count'].values.reshape(-1, 1)

# X = target.reshape(-1, 1)

y = region_prim['score'].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=100)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data

### BEGIN SOLUTION
X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train_scaled)
```

```python
predictions = model.predict(X_test_scaled)
model.fit(X_train_scaled, y_train_scaled)
plt.scatter(model.predict(X_train_scaled), y_train_scaled - model.predict(X_train_scaled), c="blue", label="Training Data")
plt.scatter(model.predict(X_test_scaled), y_test_scaled - model.predict(X_test_scaled), c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_test_scaled.min(), xmax=y_test_scaled.max())
plt.title("Residual Plot")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(y_test_scaled, predictions)
r2 = model.score(X_test_scaled, y_test_scaled)
### END SOLUTION

print(f"MSE: {MSE}, R2: {r2}")

# MSE: 0.593862095641301, R2: 0.3944770533033918
```
